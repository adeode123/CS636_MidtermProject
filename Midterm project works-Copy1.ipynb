```r
library(dplyr)
ibm<- read.csv("ibm.csv")
apple<- read.csv("aapl.csv")
amazon<- read.csv("amzn.csv")
fb<- read.csv("fb.csv")
google<- read.csv("goog.csv")
microsoft<- read.csv("msft.csv")
netflix<- read.csv("nflx.csv")
oracle<- read.csv("orcl.csv")
sap<- read.csv("sap.csv")
tesla<- read.csv("tsla.csv")
```

```r
market <- as.data.frame(cbind(Date=ibm$Date,ibmclose=ibm$Close,ibmadjclose=ibm$Adj.Close,appleclose=apple$Close,appleadjclose=apple$Adj.Close,amznclose=amazon$Close, amznadjclose=amazon$Adj.Close, fbclose=fb$Close,fbadjclose=fb$Adj.Close,googleclose=google$Close,googleadjclose=google$Adj.Close,msftclose=microsoft$Close,msftadjclose=microsoft$Adj.Close,nflxclose=netflix$Close,nflxadjclose=netflix$Adj.Close,orclclose=oracle$Close,orcladjclose=oracle$Adj.Close,sapclose=sap$Close,sapadjclose=sap$Adj.Close,tslaclose=tesla$Close,tslaadjclose=tesla$Adj.Close))
```

```r
##### Buying five different stock day 1.############
investment_amount = 5000000
cash=0
for(i in 1:nrow(market))
{
##### On day 1, buy five stocks.########
  if (i==1)
  {
    today  <- market[i,'Date']
    stocks <- c('IBM','apple','amazon','fb','google')
    ibmstockamount <- 0
    applestockamount <- 0
    amazonstockamount <- 0
    fbstockamount <- 0
    googlestockamount <- 0
    msftstockamount <- 0
    nflxstockamount <- 0
    orclstockamount <- 0
    sapstockamount <- 0
    tslastockamount <- 0
    
    ibmshares <- 0
    appleshares <- 0
    amazonshares<-0
    fbshares <-0
    googleshares<-0
    msftshares<-0
    nflxshares<-0
    orclshares<-0
    sapshares<-0
    tslashares<-0
    
    ibmstockamount <- round(floor(1000000/as.numeric(market[market$Date==today,'ibmclose']))*as.double(market[market$Date==today,'ibmclose']),2)
    cash = round(cash+(investment_amount/5)-ibmstockamount,2)
    ibmshares <- ibmshares + floor(1000000/as.numeric(market[market$Date==today,'ibmclose']))
    
    applestockamount <- round(floor(1000000/as.numeric(market[market$Date==today,'appleclose']))*as.double(market[market$Date==today,'appleclose']),2)
    cash = round(cash+(investment_amount/5)-applestockamount,2)
    appleshares <- appleshares + floor(1000000/as.numeric(market[market$Date==today,'appleclose']))
    
    amazonstockamount <- round(floor(1000000/as.numeric(market[market$Date==today,'amznclose']))*as.double(market[market$Date==today,'amznclose']),2)
    cash = round(cash+(investment_amount/5)-amazonstockamount,2)
    amazonshares <- amazonshares + floor(1000000/as.numeric(market[market$Date==today,'amznclose']))
    
    fbstockamount <- round(floor(1000000/as.numeric(market[market$Date==today,'fbclose']))*as.double(market[market$Date==today,'fbclose']),2)
    cash = round(cash+(investment_amount/5)-fbstockamount,2)
    fbshares <- fbshares + floor(1000000/as.numeric(market[market$Date==today,'fbclose']))
    
    
    googlestockamount <- round(floor(1000000/as.numeric(market[market$Date==today,'googleclose']))*as.double(market[market$Date==today,'googleclose']),2)
    cash = round(cash+(investment_amount/5)-googlestockamount,2)
    googleshares <- googleshares + floor(1000000/as.numeric(market[market$Date==today,'googleclose']))
  
  ##### Calculation for mtm.(Step 1 of the project).########## 
  
    market[i,'mtm'] <- cash+ibmshares*as.double(market[i,'ibmclose']) +
    appleshares*as.double(market[i,'appleclose']) +
    amazonshares*as.double(market[i,'amznclose']) +
    fbshares*as.double(market[i,'fbclose']) +
    googleshares*as.double(market[i,'googleclose']) +
    tslashares*as.double(market[i,'tslaclose']) +
    nflxshares*as.double(market[i,'nflxclose']) +
    msftshares*as.double(market[i,'msftclose']) +
    sapshares*as.double(market[i,'sapclose']) +
    orclshares*as.double(market[i,'orclclose']) 
  }
###### On day 6, sell lowest price stock from portfolio and buy five stocks from market(buying at #low)(step 2 of project)#####
    
  if (i == 6)
  {

    today  <- market[i,'Date']
    ibmchangeratio = (as.numeric(market[market$Date == today,'ibmadjclose']) - as.numeric(market[i-5,'ibmadjclose']))/as.numeric(market[i-5,'ibmadjclose'])*100
    
    applechangeratio = (as.numeric(market[market$Date == today,'appleadjclose']) - as.numeric(market[i-5,'appleadjclose']))/as.numeric(market[i-5,'appleadjclose'])*100
    
    amznchangeratio = (as.numeric(market[market$Date == today,'amznadjclose']) - as.numeric(market[i-5,'amznadjclose']))/as.numeric(market[i-5,'amznadjclose'])*100
    
    
    fbchangeratio = (as.numeric(market[market$Date == today,'fbadjclose']) - as.numeric(market[i-5,'fbadjclose']))/as.numeric(market[i-5,'fbadjclose'])*100
    
    googlechangeratio = (as.numeric(market[market$Date == today,'googleadjclose']) - as.numeric(market[i-5,'googleadjclose']))/as.numeric(market[i-5,'googleadjclose'])*100
    
    ratios <- c(ibmchangeratio,applechangeratio,amznchangeratio,fbchangeratio,googlechangeratio)
    
    
    
    cash = cash + applestockamount
    applestockamount=0
    
    msftchangeratio = (as.numeric(market[market$Date == today,'msftadjclose']) - as.numeric(market[i-5,'msftadjclose']))/as.numeric(market[i-5,'msftadjclose'])*100
    
    nflxchangeratio = (as.numeric(market[market$Date == today,'nflxadjclose']) - as.numeric(market[i-5,'nflxadjclose']))/as.numeric(market[i-5,'nflxadjclose'])*100
    
    orclchangeratio = (as.numeric(market[market$Date == today,'orcladjclose']) - as.numeric(market[i-5,'orcladjclose']))/as.numeric(market[i-5,'orcladjclose'])*100
    
    sapchangeratio = (as.numeric(market[market$Date == today,'sapadjclose']) - as.numeric(market[i-5,'sapadjclose']))/as.numeric(market[i-5,'sapadjclose'])*100
    
    tslahangeratio = (as.numeric(market[market$Date == today,'tslaadjclose']) - as.numeric(market[i-5,'tslaadjclose']))/as.numeric(market[i-5,'tslaadjclose'])*100
    
    ibmchangeratio
    applechangeratio #
    amznchangeratio
    fbchangeratio #
    googlechangeratio #
    msftchangeratio #
    nflxchangeratio
    orclchangeratio
    sapchangeratio #
    tslahangeratio
    
    newinvamount = cash/5
    
    applestockamount <- applestockamount + round(floor(newinvamount/as.numeric(market[market$Date==today,'appleclose']))*as.double(market[market$Date==today,'appleclose']),2)
    cash = round(cash-round(floor(newinvamount/as.numeric(market[market$Date==today,'appleclose']))*as.double(market[market$Date==today,'appleclose']),2),2)
    appleshares <- appleshares + floor(newinvamount/as.numeric(market[market$Date==today,'appleclose']))
    
    
    fbstockamount <- fbstockamount + round(floor(newinvamount/as.numeric(market[market$Date==today,'fbclose']))*as.double(market[market$Date==today,'fbclose']),2)
    cash = round(cash-round(floor(newinvamount/as.numeric(market[market$Date==today,'fbclose']))*as.double(market[market$Date==today,'fbclose']),2),2)
    fbshares <- fbshares + floor(newinvamount/as.numeric(market[market$Date==today,'fbclose']))
    
    
    msftstockamount <- msftstockamount+round(floor(newinvamount/as.numeric(market[market$Date==today,'msftclose']))*as.double(market[market$Date==today,'msftclose']),2)
    cash = round(cash-round(floor(newinvamount/as.numeric(market[market$Date==today,'msftclose']))*as.double(market[market$Date==today,'msftclose']),2),2)
    msftshares <- msftshares + floor(newinvamount/as.numeric(market[market$Date==today,'msftclose']))
    
    
    googlestockamount <- googlestockamount+round(floor(newinvamount/as.numeric(market[market$Date==today,'googleclose']))*as.double(market[market$Date==today,'googleclose']),2)
    cash = round(cash-round(floor(newinvamount/as.numeric(market[market$Date==today,'googleclose']))*as.double(market[market$Date==today,'googleclose']),2),2)
    googleshares <- googleshares + floor(newinvamount/as.numeric(market[market$Date==today,'googleclose']))
    
    
    sapstockamount <- sapstockamount+round(floor(newinvamount/as.numeric(market[market$Date==today,'sapclose']))*as.double(market[market$Date==today,'sapclose']),2)
    cash = round(cash-round(floor(newinvamount/as.numeric(market[market$Date==today,'sapclose']))*as.double(market[market$Date==today,'sapclose']),2),2)
    msftshares <- sapshares + floor(newinvamount/as.numeric(market[market$Date==today,'sapclose']))
    
    market[i,'mtm'] <- cash+ibmshares*as.double(market[i,'ibmclose']) +
    appleshares*as.double(market[i,'appleclose']) +
    amazonshares*as.double(market[i,'amznclose']) +
    fbshares*as.double(market[i,'fbclose']) +
    googleshares*as.double(market[i,'googleclose']) +
    tslashares*as.double(market[i,'tslaclose']) +
    nflxshares*as.double(market[i,'nflxclose']) +
    msftshares*as.double(market[i,'msftclose']) +
    sapshares*as.double(market[i,'sapclose']) +
    orclshares*as.double(market[i,'orclclose'])         
  }
}
### This requires optimization for it re-run after 2018-01-09 (ANYONE GOT AN IDEA HOW TO MAKE AT HAPPEN?????)
```
